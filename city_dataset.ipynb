```python
import requests
import itertools
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Replace 'YOUR_API_KEY' with your actual ipstack API key
api_key = '0f3317fb50f50cc2ddc0ea28259c8ba1'

# Function to get location information for a city
def get_location(city):
    base_url = 'http://api.ipstack.com/'
    response = requests.get(f'{base_url}{city}?access_key={api_key}')
    data = response.json()
    return data['latitude'], data['longitude']

# List of cities for which you want to calculate distances
cities = cities  # Replace with your list of cities

# Generate all pairs of cities
city_pairs = list(itertools.combinations(cities, 2))

# Initialize lists to store data
from_city = []
to_city = []
road_distances = []
flight_distances = []

# Function to calculate distance using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    # Radius of the Earth in kilometers
    radius = 6371.0

    # Convert latitude and longitude from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = radius * c

    return distance

# Calculate distances
for (city1, city2) in city_pairs:
    # Get location information for city1
    lat1, lon1 = get_location(city1)

    # Get location information for city2
    lat2, lon2 = get_location(city2)

    # Calculate road distance using Haversine formula
    road_distance = haversine(lat1, lon1, lat2, lon2)

    # Calculate flight distance (as the crow flies) using Haversine formula
    flight_distance = haversine(lat1, lon1, lat2, lon2)

    # Append data to lists
    from_city.append(city1)
    to_city.append(city2)
    road_distances.append(road_distance)
    flight_distances.append(flight_distance)

# Create a DataFrame
data = {
    'From City': from_city,
    'To City': to_city,
    'Road Distance (km)': road_distances,
    'Flight Distance (km)': flight_distances
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('city_distances.csv', index=False)
```
